Obtain the average of the following parameters by month: temperature in celsius, wind, sola

```python
data$TempC <- (data$Temp - 32) * (5/9)
print (data)
# Calcular la temperatura mÃ­nima por mes
temp_min <- aggregate(TempC ~ Month, data = data, FUN = min)

# Calcular la temperatura mÃ¡xima por mes
temp_max <- aggregate(TempC ~ Month, data = data, FUN = max)

# Combinar los resultados en un nuevo dataframe
temperaturas <- data.frame(Month = temp_min$Month, Temp_Min = temp_min$TempC, Temp_Max = temp_max$TempC)

# Imprimir los resultados
print(temperaturas)

```

Obtain the average of the following parameters by month: temperature in celsius, wind, solar radiation and ozone.

```python
# Suponiendo que tu dataframe se llama "data" y tienes columnas llamadas "Month", "TempC", "Wind", "Solar.R", "Ozone"

# Calcular los promedios por mes
promedios <- aggregate(cbind(TempC, Wind, Solar.R, Ozone) ~ Month, data = data, FUN = mean)

# Imprimir el dataframe resultante
print(promedios)

```
