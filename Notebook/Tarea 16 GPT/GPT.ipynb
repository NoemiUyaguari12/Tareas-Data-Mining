```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

 Cargar el modelo GPT-2 y el tokenizador

```python
model_name = 'gpt2-large'
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
```

Función para generar contenido creativo

```python
def generate_creative_content(prompt, max_length=150, temperature=1.2):

    input_ids = tokenizer.encode(prompt, return_tensors='pt')
    output = model.generate(input_ids, max_length=max_length, temperature=temperature, pad_token_id=tokenizer.eos_token_id)
    generated_text = tokenizer.decode(output[:, input_ids.shape[-1]:][0], skip_special_tokens=True)
    return generated_text
```

 Experimentación con diferentes prompts y configuraciones

```python
prompts = [
    "En un rincón olvidado del mundo, donde los bosques susurran secretos y los ríos guardan historias ancestrales,",
    "una pequeña aldea enfrenta una maldición que ha perdurado durante generaciones.",
    "Los habitantes viven con miedo y esperanza, esperando el día en que un héroe valiente se atreva a desafiar el destino y romper el hechizo que los aprisiona.",
    "En una metrópolis futurista gobernada por máquinas, la humanidad lucha por sobrevivir en las sombras de la tecnología"
]

for prompt in prompts:
    print(f"Prompt: {prompt}")
    print(generate_creative_content(prompt))
    print("\n" + "-"*50 + "\n")
```

```python
def user_feedback_loop():
    while True:
        initial_prompt = input("Ingrese un prompt inicial: ")
        generated_text = generate_creative_content(initial_prompt)
        print("Texto generado:")
        print(generated_text)
        user_input = input("Ingrese un prompt de continuación (o 'salir' para finalizar): ")
        if user_input.lower() == 'salir':
            break
        else:
            generated_text = generate_creative_content(user_input, max_length=100)
            print("Texto generado:")
            print(generated_text)

# Inicia el bucle de retroalimentación del usuario
user_feedback_loop()
```
