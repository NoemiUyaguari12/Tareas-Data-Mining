Select the days in May with a temperature above 70 Fahrenheit

```python

csv_path='/kaggle/input/new-york-quality/airquality.csv'
data = pd.read_csv(csv_path)
print(data.head(5))

datos_filtrados = data.loc[(data['Month'] == 5) & (data['Temp'] > 70)]


# Imprimir los datos filtrados
print("May days")
print(datos_filtrados)


```

<h3> Create a new attribute TempC which represents the temperature values in Celsius.


```python
data['TempC'] = (data['Temp'] - 32) * (5/9)
print (data)
```

<h3>Inspect which were the 30 hottest days.


```python
columna_ordenada = data['Temp'].sort_values(ascending=False)

#print(columna_ordenada.head(30))

data_mayores = data[data['Temp'].isin(columna_ordenada.head(30))]

# Imprimir el DataFrame "data_mayores"
print(data_mayores)
```

 <h3> Inspect which were the hottest days, but also with the highest ozone values.

```python
data_ordenada = data.sort_values(by=['Temp'], ascending=False)

# Seleccionar los días con los valores más altos de temperatura y ozono
dias_calurosos_temp = data_ordenada.head(30)

data_ordenada2 = data.sort_values(by=['Ozone'], ascending=False)
dias_calurosos_ozono = data_ordenada2.head(30)


# Imprimir los días más calurosos y con los valores más altos de ozono
print(dias_calurosos_temp)
print(dias_calurosos_ozono)

n = 5  # Número de registros a seleccionar
dias_calurosos_ozono = data_ordenada.nlargest(n, ['Temp', 'Ozone'])

# Imprimir los días más calurosos con los valores más altos de ozono
print(dias_calurosos_ozono)


# Imprimir el DataFrame "data_mayores"
#print(data_mayoresO)
```

<h3> Inspect the number of days for which there was a register for each month.

```python
frecuencia_meses = data['Month'].value_counts()

# Imprimir la frecuencia de cada valor
print(frecuencia_meses)
```

<h3>Obtain the average of the following parameters by month: temperature in celsius, wind, sola

```python
# Calcular la temperatura mínima por mes
temp_min = data.groupby('Month')['TempC'].min().reset_index()

# Calcular la temperatura máxima por mes
temp_max = data.groupby('Month')['TempC'].max().reset_index()

# Combinar los resultados en un nuevo DataFrame
temperaturas = pd.DataFrame({'Month': temp_min['Month'], 'Temp_Min': temp_min['TempC'], 'Temp_Max': temp_max['TempC']})

# Imprimir los resultados
print(temperaturas)

```

<h3>Obtain the average of the following parameters by month: temperature in celsius, wind, solar radiation and ozone.

```python
# Calcular los promedios por mes
promedios = data.groupby('Month').mean()[['TempC', 'Wind', 'Solar.R', 'Ozone']]

# Imprimir el DataFrame resultante
print(promedios)
```
