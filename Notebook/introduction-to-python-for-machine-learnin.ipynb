```python
csv_path='/kaggle/input/fire-forest/forestfires.csv'
df= pd.read_csv(csv_path, encoding='latin-1')
print(df)

# Contar las ocurrencias por mes
ocurrencias_por_mes = df['month'].value_counts().sort_index()

# Crear el diagrama de líneas
plt.plot(ocurrencias_por_mes.index, ocurrencias_por_mes.values, marker='o')

# Personalizar el gráfico
plt.xlabel('Mes')
plt.ylabel('Número de ocurrencias')
plt.title('Diagrama de Líneas: Ocurrencias por Mes')

# Mostrar el gráfico
plt.show()
```
