<h2> (a) Apply range-based normalization and z-score normalization.
Tip: use the function transform().

```r
library(dplyr)
data <-read.csv("/kaggle/input/cars-database/imports-85.data",, header = FALSE, na.strings = "?",
                 col.names = c("SYMBOLING", "NORMALIZED-LOSSES", "MAKE", "FUEL-TYPE", "ASPIRATION",
                               "NUM-OF-DOORS", "BODY-STYLE", "DRIVE-WHEELS", "ENGINE-LOCATION", "WHEEL-BASE",
                               "LENGTH", "WIDTH", "HEIGHT", "CURB-WEIGHT", "ENGINE-TYPE", "NUM-OF-CYLINDERS",
                               "ENGINE-SIZE", "FUEL-SYSTEM", "BORE", "STROKE", "COMPRESSION-RATIO",
                               "HORSEPOWER", "PEAK-RPM", "CITY-MPG", "HIGHWAY-MPG", "PRICE"))

# Cargar el dataSet
data <- data.frame(data)
head(data,10)
colname=colnames(data)
print (colname)


#Imputar los valores faltantes con 0

data_imputed <- replace(data, is.na(data), 0)

head(data_imputed, 10)

# Normalización basada en rango
normalized_data_range <- transform(data_imputed,
                                   variable1_range = (PRICE - min(PRICE)) / (max(PRICE) - min(PRICE)))

# Normalización de puntuación z
normalized_data_zscore <- transform(data_imputed,
                                    variable1_zscore = (PRICE - mean(PRICE)) / sd(PRICE))

head(normalized_data_range,10)
head(normalized_data_zscore,10)
```

<h3> (b) Discretize it into 4 equal-frequency ranges an into 4 equal-width ranges.

```r
# Calcula los cortes para rangos de igual frecuencia
cuts_frequency <- quantile(data$PRICE, probs = seq(0, 1, length.out = 5), na.rm = TRUE)

# Discretiza la variable PRICE en rangos de igual frecuencia
data$discrete_frequency <- cut(data$PRICE, breaks = cuts_frequency, labels = FALSE)

# Verificar los rangos resultantes
table(data$discrete_frequency)

# Crear 4 rangos de igual ancho utilizando la función cut()
data$discrete_width <- cut(data$PRICE, breaks = 4, labels = FALSE, include.lowest = TRUE)

# Verificar los rangos resultantes
table(data$discrete_width)



```

<h3> 4. Load the package corrplot and select the numeric attributes of the car insurance data set.
    
(a) Using the function cor(), obtain the pearson correlation coefficient between each pair of variables.
(b) Apply the function cor.mtest() to the previous result to calculate the p-values and confidence
intervals of the correlation coefficient for each pair of variables.
(c) Plot the all correlation information using the function corrplot. Explore some of its parameters

```r
library(corrplot)
library(dplyr)
data <-read.csv("/kaggle/input/cars-database/imports-85.data",, header = FALSE, na.strings = "?",
                 col.names = c("SYMBOLING", "NORMALIZED-LOSSES", "MAKE", "FUEL-TYPE", "ASPIRATION",
                               "NUM-OF-DOORS", "BODY-STYLE", "DRIVE-WHEELS", "ENGINE-LOCATION", "WHEEL-BASE",
                               "LENGTH", "WIDTH", "HEIGHT", "CURB-WEIGHT", "ENGINE-TYPE", "NUM-OF-CYLINDERS",
                               "ENGINE-SIZE", "FUEL-SYSTEM", "BORE", "STROKE", "COMPRESSION-RATIO",
                               "HORSEPOWER", "PEAK-RPM", "CITY-MPG", "HIGHWAY-MPG", "PRICE"))


# Seleccionar los atributos numéricos
numeric_data <- data[, sapply(data, is.numeric)]

# Calcular el coeficiente de correlación de Pearson utilizando la función cor()
cor_matrix <- cor(numeric_data, use = "pairwise.complete.obs")

# Aplicar la función cor.mtest() para calcular los valores p y los intervalos de confianza
cor_test <- cor.mtest(numeric_data)

# Graficar la matriz de correlación utilizando la función corrplot()
corrplot(cor_matrix, method = "color", type = "upper", order = "hclust", tl.col = "black", 
         tl.srt = 45, addCoef.col = "black", number.cex = 0.7)
```
