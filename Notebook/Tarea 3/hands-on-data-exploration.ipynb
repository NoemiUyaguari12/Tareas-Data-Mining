<h3> 1. Using the package dplyr, answer the following questions:
(a) Obtain the number of cars by bodyStyle.

```r
# Instalar y cargar el paquete dplyr
if (!requireNamespace("dplyr", quietly = TRUE)) {
  install.packages("dplyr")
}
library(dplyr)

load ("/kaggle/input/carsfinal/carIns_final (2).Rdata")
data <- as.data.frame(carIns_final)
head (carIns_final,10)

#Agrupar 
data_agrupada <- data %>%
group_by(bodyStyle) %>%
summarise(cantidad = n())

head(data_agrupada,10)
```

(b) Obtain the number of cars by bodyStyle and fuelType

```r
data_agrupada <- data %>%
group_by (bodyStyle , fuelType) %>%
summarise( cantidad =n(), .groups = "drop")
head (data_agrupada)
```

<h3> (c) Obtain the mean and the standard deviation of the attribute cityMpg by bodyStyle in ascending order

```r
data_agrupada <- data %>%
group_by(bodyStyle) %>%
summarise (mean (cityMpg), desviacion = sd (cityMpg) )
head (data_agrupada)
```

<h3> (d) Also by bodyStyle, and for the attributes cityMpg and highwayMpg, obtain the mean, the standard deviation, the median and the inter-quartile range

```r
data_agrupada <- data %>%
group_by (bodyStyle) %>%
summarise (Media= mean(cityMpg), desviacion = sd (cityMpg), mediana = median(cityMpg), 
           rango= IQR(cityMpg), .groups="drop" )
head (data_agrupada)

data_agrupada <- data %>%
group_by (bodyStyle ) %>%
summarise (Media= mean(highwayMpg), desviacion = sd (highwayMpg), mediana = median(highwayMpg), rango= IQR(highwayMpg), 
        )
head (data_agrupada)
```

<h3> (g) Show the distribution of cars by price. Suggestion: create bins of width equal to 5000.

<h3> (i) Check (visually) if it is plausible to consider that price follows a normal distribution

```r
library(ggplot2)
#Cargar el data set
load("/kaggle/input/carsfinal/carIns_final (2).Rdata")
data <- as.data.frame(carIns_final)
head(data, 10)
media <- mean(data$price)
#Dibujar un grafico de densidad 
ggplot(data, aes(x = price)) +
  geom_density() +
  labs(x = "Precio", y = "Densidad", title = "Distribución de precios") +
  stat_function(fun = dnorm, args = list(mean = mean(data$price), sd = sd(data$price)), color = "red", lwd = 1)
print ('Graficamente el precio sigue un distibución normal')

```
